```python
y=data['RainToday']
y.map({'Yes':1,'No':0})
```

```python
clean_data=data[['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
       'WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm',
       'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am',
       'Temp3pm', 'RISK_MM']]
clean_data.head()
```

```python
clean_data['WindGustDir'].unique()     # unique values of directions
```

```python
directions=clean_data['WindGustDir'].unique()
```

```python
directions
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_x=LabelEncoder()
clean_data['WindGustDir']=labelencoder_x.fit_transform(clean_data['WindGustDir'])
clean_data['WindDir9am']=labelencoder_x.fit_transform(clean_data['WindDir9am'])
clean_data['WindDir3pm']=labelencoder_x.fit_transform(clean_data['WindDir3pm'])
clean_data[['WindGustDir','WindDir9am','WindDir3pm']]
```

```python
clean_data
```

```python
x_train,x_test,y_train,y_test=train_test_split(clean_data,y,test_size=0.4,random_state=30)
```

```python
x_train.shape
```

```python
humidity_classifier=DecisionTreeClassifier()
```

```python
humidity_classifier.fit(x_train,y_train)
y_test
```

```python
y_pre=humidity_classifier.predict(x_test)
accuracy_score(y_test,y_pre)
```
