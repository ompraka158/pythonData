```python
import pymysql
from tkinter import *
from tkinter import messagebox
def logout(home):
    home.destroy()
    splash()
def connection():
    db=pymysql.connect('localhost','root','','emp4')
    cursor=db.cursor()
    return cursor
def loggin(x,y,root):
    query="Select count(*) from login where username='%s' and password='%s'"%(x.get(),y.get())
    cursor=connection()
    try:
        cursor.execute(query)
        result=cursor.fetchone()
        cursor.close()
        if result[0]==1:
            root1=Tk()
            root1.title("Main")
            root1.iconbitmap('logo.ico')
            Label(root1,text='Welcome to our SYSTEM!!!').grid(row=0,column=1)
            root.destroy()
            Button(root1,text='Logout',command=lambda:logout(root1)).grid(row=1,column=2)
            root1.geometry("500x500")
            root1.mainloop()
        else:
            messagebox.showinfo("Login Failed!!! Try Again...")
    except:
        messagebox.showinfo("Login Failed!!! Try Again...")
def login(splash):
    splash.destroy()
    root=Tk()
    root.title("Login Page")
    root.iconbitmap('logo.ico')
    try:
        Label(root,text='Welcome').grid(row=0,column=2)
        Label(root,text='Username :').grid(row=1,column=1)
        e1=Entry(root)
        Label(root,text='Password :').grid(row=2,column=1)
        e2=Entry(root)
        e1.grid(row=1,column=2)
        e2.grid(row=2,column=2)
        Button(root,text='Login',command=lambda:loggin(e1,e2,root)).grid(row=3,column=2)
        root.geometry("500x500")
        root.mainloop()
    except:
        messagebox.Message("Connection Failed")
        root.destroy()
def registering(x,y,root):
    query="Select count(*) from login where username='%s' and password='%s'"%(x.get(),y.get())
    cursor=connection()
    try:
        cursor.execute(query)
        result=cursor.fetchone()
        if result[0]==0:
            query1="Insert into login values('%s','%s')"%(x.get(),y.get())
            cursor.execute(query1)
            cursor.close()
            login(root)
        else:
            messagebox.showinfo("Login Failed!!! Try Again...")
    except:
        messagebox.showinfo("Login Failed!!! Try Again...")
def register(splash):
    splash.destroy()
    root=Tk()
    root.title("Register Page")
    root.iconbitmap('logo.ico')
    try:
        Label(root,text='Welcome').grid(row=0,column=2)
        Label(root,text='Username :').grid(row=1,column=1)
        e1=Entry(root)
        Label(root,text='Password :').grid(row=2,column=1)
        e2=Entry(root)
        e1.grid(row=1,column=2)
        e2.grid(row=2,column=2)
        Button(root,text='Register',command=lambda:registering(e1,e2,root)).grid(row=3,column=2)
        root.geometry("500x500")
        root.mainloop()
    except:
        messagebox.Message("Connection Failed")
        root.destroy()
def splash():
    splash=Tk()
    splash.title("Splash")
    splash.iconbitmap('logo.ico')
    splash.geometry("500x500")
    Label(splash,text='Welcome!!!').grid(row=0,column=2)
    Button(splash,text='Login',command=lambda:login(splash)).grid(row=1,column=1)
    Button(splash,text='Register',command=lambda:register(splash)).grid(row=1,column=2)
    splash.mainloop()
splash()
```
