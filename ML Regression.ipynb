```python
#Replacing 6 with 5 and 5 with 6
x=int(input("Enter a number: "))
i=0
sum=0
while(x>0):
    r=x%10
    if(r==5):
        r=6
    elif r==6:
        r=5
    if(i>0):
        sum=(r*(10**i))+sum
    else:
        sum=r
    i=i+1
    x=int(x/10)
print (sum)
```
