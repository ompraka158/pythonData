```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
#make_blobs()[0]
```

```python
#make_blobs()[1]
```

```python
data = make_blobs(n_samples=500, n_features=2, centers=4, random_state=50)
```

```python
data[0][:,0].shape
```

```python
data[0][:,1].shape
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow') #data[0][:,0----x]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
Y_pred=kmeans.predict(data[0])
```

```python
Y_pred
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=Y_pred, cmap='rainbow') #data[0][:,0----x]
```

```python
kmeans.cluster_centers_ #center of the clusters are found
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c = kmeans.labels_)
ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c = data[1],cmap = 'rainbow')
```
